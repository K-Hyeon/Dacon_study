## 4.2 배열의 생성

### 1. 리스트에서 행렬/배열 생성

```python
import numpy as np # numpy 패키지를 로드하여 np로 사용
a = [[1,2,3], [4,5,6]] # 리스트에서 행렬 생성
b = np.array(a)
print(b)
```

```python
print(a)
```

```python
print(b.ndim) # 배열의 열수(차원)
```

```python
print(b.shape) # 배열의 차원
```

```python
print(b[0,0])  # 배열의 원소 접근
```

### 2. 특수한 배열의 생성

```python
print(np.arange(10)) # 1씩 증가하는 1차원 배열(시작이 0부터)
```

```python
print(np.arange(5, 10)) # 1씩 증가하는 1차원 배열(시작이 5부터)
```

```python
print(np.zeros((2,2)))  # 영행렬 생성
```

```python
print(np.ones((2,3)))  # 유닛행렬
```

```python
print(np.full((2,3), 5)) # 모든 원소가 5인 2*3행렬
```

```python
print(np.eye(3)) # 단위행렬
```

### 3. 배열의 차원 변환

```python
b=a.reshape((4,5))
print(b)
```

## 4.3 Numpy 슬라이싱/인덱싱

## 4.4 Numpy 연산
- 배열간 연산 가능
- +,-,*,/ 등의 연산자 사용가능
- add(),substract(),multiply(),divide()

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

# 각 요소 더하기
c = a+b
# c= np.add(a, b)

print(c)
```

```python
# 각 요소 곱하기
c = a*b
# c = np.multiply(a, b)

print(c)
```

```python
# 각 요소 나누기
# c = a/b
c = np.divide(a,b)
print(c) # [0.25 0.4 0.5]
```

- dot()함수: 행렬의 곱

```python
import numpy as np
arr1 = [[1,2],[3,4]]
arr2 = [[5,6],[7,8]]
a = np.array(arr1)
b = np.array(arr2)

c= np.dot(a, b)
print(c)
```

- 모든 원소의 합: sum(), 모든 원소의 곱: prod()

```python
import numpy as np

a = np.array([[-1,2,3],[3,4,8]])
s = np.sum(a)
print('sum=',a.sum()) 

# 행별/열별 연산 (axis=0/1)
```

```python
print('sum by row=',a.sum(axis=0))
```

```python
print('mean=',a.mean())
```

```python
# 표준편차
print('sd=',a.std())
```

```python
# 모든 요소의 곱
print('product=',a.prod()) 
```

### 1. 벡터연산(내적)

image.png

x = (1,2,3)', y = (6,23,-1)'

x'y = 1x6 + 2x23 + 3x(-1) = ?

```python
import numpy as np
x = np.array([1, 2, 3])
print(x)
```

```python
y = np.array([6, 23,-1])
print(y)
```

```python
print(x.dot(y.T))# x'y
```

```python
print(x.T.dot(y) )# x'y
```

### 3. 역행렬

```python
import numpy as np
from numpy.linalg import inv

# 난수 생성성
# np.random.seed()

X = np.array([[1,2,3],[1,0,0],[0,0,1]])
print(X)
```

```python
# 역행렬 inv
Y = inv(X)
print(Y)
```

```python
X.dot(Y)
```
