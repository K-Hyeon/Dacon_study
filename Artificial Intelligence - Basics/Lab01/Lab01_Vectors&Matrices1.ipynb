```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

# 실습 1

```python
def cal_distance(x, y):
    return abs((x[0]-x[1])**2 + (y[0]-y[1])**2)

x = np.array([-3, 2])
y = np.array([3, 2])
distance = cal_distance(x, y)

# plot 코드를 작성하시오

print(f'두 벡터 사이의 거리는 {distance}')
```

### ppt와 최대한 비슷하게 구현하였습니다.

```python
# 벡터 x와 y를 시각화
plt.figure(figsize=(8, 6))
plt.plot([0, x[0]], [0, x[1]], label='vector x', color='red', linestyle='-', linewidth=2)
plt.plot([0, y[0]], [0, y[1]], label='vector y', color='blue', linestyle='-', linewidth=2)

# 벡터 사이의 거리 시각화(초록색 점선)
plt.plot([x[0], y[0]], [x[1], y[1]], linestyle='--', color='green', label='||x-y||', marker='o')

# 벡터의 방향 화살표를 추가하고 화살표 크기 조절(mutation_scale)
arrow_style = dict(arrowstyle='-|>', color='red', mutation_scale=20)
plt.annotate('', xy=(x[0], x[1]), xytext=(0, 0), arrowprops=arrow_style)
arrow_style = dict(arrowstyle='-|>', color='blue', mutation_scale=20)
plt.annotate('', xy=(y[0], y[1]), xytext=(0, 0), arrowprops=arrow_style)

# x=0과 y=0에 검은색 점선 추가
plt.axvline(x=0, color='gray', linestyle='--')
plt.axhline(y=0, color='gray', linestyle='--')

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()

plt.title('Visualize the distance of two vectors')
plt.show()
```

# 실습 2

```python
def cal_cos_theta(x, y):
    dot_product = np.dot(x, y)
    norm_x = np.linalg.norm(x)
    norm_y = np.linalg.norm(y)
    cos_theta = dot_product / (norm_x * norm_y)
    return cos_theta

x = np.array([-3, 2])
y = np.array([3, 2])

# x와 y 단위 벡터 변환
x_unit = x / np.linalg.norm(x)
y_unit = y / np.linalg.norm(y)

cos_theta_1 = cal_cos_theta(x, y)
cos_theta_2 = cal_cos_theta(x_unit, y_unit)

# plot 코드를 작성하시오
print(f'x, y 사이의 cos theta 값은 {cos_theta_1}')
print(f'x_, y_ 사이의 cos theta 값은 {cos_theta_2}')
```

```python
# 벡터 x와 y를 시각화
plt.figure(figsize=(8, 6))
plt.plot([0, x[0]], [0, x[1]], label='vector x', color='red', linestyle='-', linewidth=2, zorder=1)
plt.plot([0, y[0]], [0, y[1]], label='vector y', color='blue', linestyle='-', linewidth=2, zorder=1)
plt.plot([0, x_unit[0]], [0, x_unit[1]], label='vector x_', color='pink', linestyle='-', linewidth=2, zorder=2)
plt.plot([0, y_unit[0]], [0, y_unit[1]], label='vector y_', color='cyan', linestyle='-', linewidth=2, zorder=2)

# 벡터의 방향 화살표를 추가하고 화살표 크기 조절(mutation_scale)
arrow_style = dict(arrowstyle='-|>', color='red', mutation_scale=20)
plt.annotate('', xy=(x[0], x[1]), xytext=(0, 0), arrowprops=arrow_style)
arrow_style = dict(arrowstyle='-|>', color='blue', mutation_scale=20)
plt.annotate('', xy=(y[0], y[1]), xytext=(0, 0), arrowprops=arrow_style)

arrow_style = dict(arrowstyle='-|>', color='pink', mutation_scale=20)
plt.annotate('', xy=(x_unit[0], x_unit[1]), xytext=(0, 0), arrowprops=arrow_style)
arrow_style = dict(arrowstyle='-|>', color='cyan', mutation_scale=20)
plt.annotate('', xy=(y_unit[0], y_unit[1]), xytext=(0, 0), arrowprops=arrow_style)

# x=0과 y=0에 검은색 점선 추가
plt.axvline(x=0, color='gray', linestyle='--')
plt.axhline(y=0, color='gray', linestyle='--')

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# legend를 왼쪽 아래에 위치하도록 
plt.legend(loc='lower left')
plt.grid()

plt.title('Visualize the distance of two vectors')
plt.show()
```

# 실습 3

```python
def projection(origin_v, target_v):
    proj_scalar = np.dot(origin_v, target_v) / np.dot(target_v, target_v)
    proj_v = proj_scalar * target_v
    return proj_v

origin_v = np.array([3, 4])
target_v = np.array([4, 0])
proj_v = projection(origin_v, target_v)

# plot 코드를 작성하시오
print(f'origin_vector:', origin_v)
print(f'target_vector:', target_v)
print(f'projected vector: ', proj_v)
```

```python
# origin_v, target_v, projected vector(proj_v) 시각화
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, origin_v[0], origin_v[1], angles='xy', scale_units='xy', scale=1, color='red', label='origin vector')
plt.quiver(0, 0, target_v[0], target_v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='target vector')
plt.quiver(0, 0, proj_v[0], proj_v[1], angles='xy', scale_units='xy', scale=1, color='green', label='projected vector')

# x=0과 y=0에 검은색 점선 추가
plt.axvline(x=0, color='gray', linestyle='--')
plt.axhline(y=0, color='gray', linestyle='--')

plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.title('Vector Projection Visualization')

# 그래프를 표시합니다.
plt.show()
```
